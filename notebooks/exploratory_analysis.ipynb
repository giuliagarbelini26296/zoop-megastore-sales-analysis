```python
import pandas as pd
import matplotlib.pyplot as plt

vendas = pd.read_csv("vendas.csv")

print(vendas.isnull().sum())
print(vendas.duplicated().sum())

vendas["data_venda"] = pd.to_datetime(vendas["data_venda"])

vendas["faturamento"] = (
    vendas["valor_venda"] *
    vendas["quantidade"]
)

vendas["mes"] = vendas["data_venda"].dt.month

fat_mes = vendas.groupby("mes")["faturamento"].sum()

fat_categoria = (
    vendas.groupby("categoria")["faturamento"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))
plt.plot(fat_mes.index, fat_mes.values, marker="o")
plt.title("Faturamento por Mês")
plt.xlabel("Mês")
plt.ylabel("Faturamento")
plt.grid()
plt.show()

plt.figure(figsize=(8,5))
fat_categoria.plot(kind="bar")
plt.title("Faturamento por Categoria")
plt.xlabel("Categoria")
plt.ylabel("Faturamento")
plt.show()
```
